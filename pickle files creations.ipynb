# Creating USA's (1) Son-Parent Dict, (2) Category ID to Name Dict, (3) Categories per Indent Dict:

```python
parent_dict = {}
weight_dict = {}
cat_cat_id_dict = {}
indent_dict = {}

parent_son_df = data[['Category', 'Category_id','Parent', 'Parent_ID']].drop_duplicates()
cat_weight_df = data[['Category_id','Weight']].drop_duplicates()
cat_cat_id_df = data[['Category_id','Category']].drop_duplicates()
indent_df = data[['Indent','Category_id']].drop_duplicates()


for category_id in list(parent_son_df['Category_id'].unique()):
    parent_dict[category_id] = parent_son_df[parent_son_df['Category_id'] == category_id]['Parent_ID'].values[0]
    
for category_id in list(cat_weight_df['Category_id'].unique()):
    weight_dict[category_id] = cat_weight_df[cat_weight_df['Category_id'] == category_id]['Weight'].values[0]
    
for category_id in list(cat_cat_id_df['Category_id'].unique()):
    cat_cat_id_dict[category_id] = cat_cat_id_df[cat_cat_id_df['Category_id'] == category_id]['Category'].values[0]
    
for indent in list(indent_df['Indent'].unique()):
    indent_dict[indent] = list(indent_df[indent_df['Indent']==indent].Category_id.unique())
    
```

```python
with open(usa_parent_son_pickle, 'wb') as handle: 
    pickle.dump(parent_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open(usa_cat_weight_pickle, 'wb') as handle: 
    pickle.dump(weight_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
with open(usa_cat_id_to_name_pickle, 'wb') as handle: 
    pickle.dump(cat_cat_id_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open(usa_cats_per_indent_pickle, 'wb') as handle: #wb means write - creating a pickle file
    pickle.dump(indent_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# print categories with no parents - making sure the only category without parent id is all items (8106)
for cat_id in list(parent_dict.keys()):
    if pd.isna(parent_dict[cat_id]):
        print(cat_id)
```

```python
parent_to_sons_dict = {}

for cat_id in list(data.Category_id.unique()):
    if parent_dict[cat_id] in list(parent_to_sons_dict.keys()):
        parent_to_sons_dict[parent_dict[cat_id]]['sons'].append(cat_id)
    
    else:
        parent_to_sons_dict[parent_dict[cat_id]] = {'sons': [cat_id]}
```

```python
reversed_parent_son_corr_dict = {}

for cat_id in list(pivot_corr_df.Category_id.unique()):
    relevant_columns = [x for x in pivot_corr_df.columns if x not in ['Date', 'Category_id']]
    son_array = pivot_corr_df[(pivot_corr_df['Category_id']==cat_id)][relevant_columns].values.flatten().tolist()
    parent_array = pivot_corr_df[(pivot_corr_df['Category_id']==parent_dict[cat_id])][relevant_columns].values.flatten().tolist()
    if parent_dict[cat_id] not in pivot_corr_df.Category_id.unique():
        corr = None
    else:
        corr = stats.pearsonr(son_array,parent_array)[0]
    
    if parent_dict[cat_id] in list(reversed_parent_son_corr_dict.keys()):
        reversed_parent_son_corr_dict[parent_dict[cat_id]]['sons'].append(cat_id)
        reversed_parent_son_corr_dict[parent_dict[cat_id]]['corrs'].append(corr)
    
    else:
        reversed_parent_son_corr_dict[parent_dict[cat_id]] = {'sons': [cat_id], 'corrs': [corr]}
```

```python
reversed_parent_son_dict = {}
reversed_parent_son_corr_final = {} 

for parent_id in list(reversed_parent_son_corr_dict.keys()):
    corrs = reversed_parent_son_corr_dict[parent_id]['corrs']
    if None in corrs:
        reversed_parent_son_dict[parent_id] = reversed_parent_son_corr_dict[parent_id]['sons'][0]
        reversed_parent_son_corr_final[(parent_id, reversed_parent_son_corr_dict[parent_id]['sons'][0])] = None
    else:
        argmax_corr = np.argmax(corrs)
        highest_corr_son = reversed_parent_son_corr_dict[parent_id]['sons'][argmax_corr]
        
        reversed_parent_son_dict[parent_id] = highest_corr_son
        reversed_parent_son_corr_final[(parent_id, highest_corr_son)] = reversed_parent_son_corr_dict[parent_id]['corrs'][argmax_corr]
```
